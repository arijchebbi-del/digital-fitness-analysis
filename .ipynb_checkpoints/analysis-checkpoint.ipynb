# Digital Fitness Market Analysis
## Objective
Understand global and national trends in digital fitness classes using Google Trends data.

## Step 1: Import Libraries
We will use Pandas for data manipulation and Matplotlib for visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2: Load Data
Read CSV files into Pandas DataFrames.

```python
workoutgeo = pd.read_csv("data/workout_geo.csv", index_col=0)
threegeo = pd.read_csv("data/three_keywords_geo.csv", index_col=0)
three = pd.read_csv("data/three_keywords.csv")
workout = pd.read_csv("data/workout.csv")

# Preview the data
workout.head()
three.head()
workoutgeo.head()
threegeo.head()
```

## Step 3: Worldwide Trends
Plot the popularity of different fitness keywords over time.

```python
# Workout worldwide
plt.figure(figsize=(60,4))
plt.plot(workout["month"], workout["workout_worldwide"])
plt.title("Workout Worldwide Popularity Over Time")
plt.xlabel("Month")
plt.ylabel("Popularity Index")
plt.show()

# Home workout
plt.figure(figsize=(60,4))
plt.plot(three["month"], three['home_workout_worldwide'])
plt.title("Home Workout Worldwide Popularity")
plt.show()

# Gym workout
plt.figure(figsize=(60,4))
plt.plot(three["month"], three['gym_workout_worldwide'])
plt.title("Gym Workout Worldwide Popularity")
plt.show()

# Home & Gym
plt.figure(figsize=(60,4))
plt.plot(three["month"], three['home_gym_worldwide'])
plt.title("Home & Gym Worldwide Popularity")
plt.show()
```

## Step 4: Country-Level Insights
Compare popularity of workouts in different countries.

```python
# United States, Australia, Japan
print("Workout popularity (2018–2023):")
print("United States:", workoutgeo.loc['United States'])
print("Australia:", workoutgeo.loc['Australia'])
print("Japan:", workoutgeo.loc['Japan'])

# Philippines, Malaysia
print("\nHome/Gym trends (2018–2023):")
print("Philippines:", threegeo.loc['Philippines'])
print("Malaysia:", threegeo.loc['Malaysia'])
```

## Step 5: Observations & Insights
- The United States had the highest overall interest in workouts.
- Australia had moderate interest, while Japan was low.
- Home workouts peaked in the Philippines, showing strong at-home fitness trends.
- Gym workouts remain popular in Malaysia, suggesting continued demand for gyms.
- Trends vary across countries, which can guide product strategy for digital fitness services.
